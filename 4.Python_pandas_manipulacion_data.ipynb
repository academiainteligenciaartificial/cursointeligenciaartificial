# Manipulación de datos

```python
import pandas as pd
```

```python
dataset = pd.read_csv("sample_data/california_housing_train.csv")
dataset.head()
```

## Agregar una columna

```python
dataset["population2"] = dataset["population"].apply(lambda x: x * 2)
```

```python
dataset
```

## Eliminar una columna

```python
dataset.drop("population2", axis=1, inplace=True)
```

```python
dataset
```

## Obteniendo una muestra del dataset

```python
dataset80 = dataset.sample(frac=0.8)
dataset80
```
