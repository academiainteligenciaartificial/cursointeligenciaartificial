# Funciones lambda anonimas

```python
lista = [3, 4, 5, 6, 7, 8, 9]
lista
```

```python
def multiplicar_por_dos(valor): 
  return valor * 2
```

```python
multiplicar_por_dos(2)
```

```python
nueva_lista = []
for valor in lista:
  nueva_lista.append(multiplicar_por_dos(valor))
```

```python
nueva_lista
```

```python
nueva_lista2 = list(map(multiplicar_por_dos, lista))
nueva_lista2
```

```python
nueva_lista3 = list(map(lambda x: x * 3, lista))
nueva_lista3
```
